```python
def first(lst):
    try:
        return lst[0]
    except IndexError:
        return lst
    
def chem_lookup(chem_id):
    return first(chems[chems['id'] == chem_id]['name'].values)

def rgroup_lookup(chem_id):
    r_id = first(mcr[mcr['chem_id'] == chem_id]['react_id'].values)
    return reacts.loc[r_id]['name']

def iupac_lookup(chem_id):
    #first(chems[chems['id'] == chem_id]['name'].values)
    pass # idk where this is yet

def ids_by_name(namePartial, asmask=False):
    idx = chems['name'].str.find(namePartial.upper())
    if asmask:
        return idx > -1
    return chems['id'][idx > -1]

def hazards_by_name(namePartial):
    ids = ids_by_name(namePartial)
    df = pd.concat([ids.apply(chem_lookup), ids.apply(rgroup_lookup)], axis=1)
    df.columns = ['name', 'react_group']
    return df
```
